```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('CALICO Fall \'24 Registrations - Form Responses.csv')

# Drop the second row (index 1)
df = df.drop(index=0) # Drops the row at index 0 (the first row with team info)

df.tail(10)
```

```python
# Create list of Team IDs
starting_id = 3900
num_teams = len(df)
team_ids = range(3900, 3900 + num_teams)

# Create accounts/teams only for new registrations
# drop columns that have already been done
# batch 1: index 1-388
## df.iloc[[389:]] # for batch 2
```

```python
### CREATING TEAMS

# Extract the 'Team Name' and 'Organization' columns
team_names = df['Team Name']
organizations = df['Organization']

# Create a dictionary to build the DataFrame – better for this use case
data = {
    'File_Version': team_ids,
    '2': '',
    'Unnamed: 2': 21,  # Give a unique temporary name
    'Team Name': team_names,
    'Organization 1': organizations,
    'Organization 2': organizations,
    'Unnamed: 6': '',  # Temporary names for blank columns
    'Unnamed: 7': '',
    'Placeholder': 'placeholder'
}

teams_df = pd.DataFrame(data)

# Remove the temporary headers that should be blank. This *must* be done after the DataFrame is created
teams_df = teams_df.rename(columns={'Unnamed: 2': '', 'Team Name': '', 'Organization 1': '', 'Organization 2': '', 'Unnamed: 6': '', 'Unnamed: 7': '', 'Placeholder': ''})

teams_df.head(20)
```

```python
### CREATING ACCOUNTS

import random

# Method to generate passwords
def make_random_oreo_case(tokens, max_length):
    def random_token(n):
        t = ["O", "RE", "&"]
        first = random.randint(0, 1)
        return t[first] + ''.join(random.choices(t, k=n-1))
    pwd = random_token(tokens)
    while len(pwd) > 80:
      pwd = random_token(tokens)
    return pwd # Return just the string, not the TestCase

# Create the usernames and passwords lists
usernames = []
passwords = []

for team_id in team_ids:
    username = ''.join(random.choices('abcdefghijklmnopqrstuvwxyz', k=8)) + str(team_id)
    password = make_random_oreo_case(42, 80)
    usernames.append(username)
    passwords.append(password)

# Create the accounts DataFrame
accounts_df = pd.DataFrame({
    'accounts': ['team'] * num_teams,
    '1': '',
    'Username': usernames,   # No header, add usernames
    'Password': passwords #No header, add passwords
})

# Remove temporary header and rename columns for correct output
accounts_df = accounts_df.rename(columns={'Username': '', 'Password': ''})

accounts_df.head(20)
```

```python
# Save to TSV
teams_df.to_csv('teams.tsv', sep='\t', index=False)
accounts_df.to_csv('accounts.tsv', sep='\t', index=False)
```
